```python
import pandas as pd

df = pd.read_csv("data/master/nahr_ibrahim_master_model.csv", parse_dates=["date"])

# Annual cycle: mean per month across all years, for both soil moisture layers
df["month"] = df["date"].dt.month

print("Monthly climatology (mean across all years):\n")
print("Month  sm_28_100cm (shallow main)  sm_100_289cm (deep, raw values)")
shallow = df.groupby("month")["soil_moisture_mm"].mean()
# we need to bring sm_100_289cm_mm back — it's not in the model CSV by name,
# only sm_deep_30day is. Use that instead.
deep = df.groupby("month")["sm_deep_30day"].mean()

for m in range(1, 13):
    print(f"  {m:2d}   {shallow[m]:>20.1f}  {deep[m]:>20.1f}")

print("\nPrecipitation climatology for context:")
precip = df.groupby("month")["precip_mm_day"].mean()
for m in range(1, 13):
    print(f"  {m:2d}   {precip[m]:>8.2f} mm/day")
```
